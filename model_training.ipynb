#### **Download "ultralytics" library to load YOLO (You Only Look Once) model.**

```python
# import YOLOv8n model from ultralytics
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
# model training upto 100 epochs, and image size is (480 X 480)

model.train(data = '/kaggle/input/datasets/partar/yaml-data/data.yaml', epochs = 100, imgsz = 480)
```

```python
# Evaluation of model.
metrics = model.val(data = '/kaggle/input/datasets/partar/yaml-data/data.yaml', split = 'test')

print("\nTest Results:")
print(f"mAP50: {metrics.box.map50}")
print(f"mAP50-95: {metrics.box.map}")
print(f"Mean Precision: {metrics.box.mp}")
print(f"Mean Recall: {metrics.box.mr}")
```

```python
# Testing the model performance.
result = model.predict(source = '/kaggle/input/datasets/partar/aerial-object/test/test/images/00188d7f40a84793_jpg.rf.7f9da2b662dc236fbdcc1f22d8e0983e.jpg',
             conf = 0.5,
             save = False)
result[0].show()
```
